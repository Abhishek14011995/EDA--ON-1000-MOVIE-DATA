# **8. Exploratory Data Analysis**
---

- This section is emphasised on asking the right questions and perform analysis using the data.

- Note that there is no limit how deep you can go, but make sure not to get distracted from right track.

# Number of movies released in A year

```python
figure = plt.figure(figsize=[15, 7])
data['Year'].value_counts().sort_index().plot.bar(color="#f8c471",grid=True)
plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 350, 50), size=12)
plt.xlabel(xlabel='YEARS', size=14)
plt.ylabel(ylabel='COUNT', size=14)
plt.title(label='Number of movies released in A year', size=16)
plt.show()
```

# Top 10 rowa of Genre

```python
figure = plt.figure(figsize=[15, 7])
data["Genre"].value_counts()[0:10].plot.bar(color='#1ABC9C',)
plt.xticks(size=12, rotation=45)
plt.yticks(ticks=np.arange(0, 70, 10), size=12)
plt.xlabel(xlabel='Genre', size=14)
plt.ylabel(ylabel='Count ', size=14,rotation=90)
plt.title(label='Bar Chart For Top 10 Genre', size=16)
plt.show()
```

# Top 10 Director With highest movies 

```python
figure = plt.figure(figsize=[15, 7])
data["Director"].value_counts()[0:10].plot.bar(color='#7d3c98',)
plt.xticks(size=12, rotation=45)
plt.yticks(ticks=np.arange(0, 12, 2), size=12)
plt.xlabel(xlabel='Director', size=14)
plt.ylabel(ylabel='Count ', size=14,rotation=90)
plt.title(label='Bar Chart For Top 10 Director With highest movies ', size=16)
plt.show()
```

# Top 10 Director With highest Revenue

```python
figure = plt.figure(figsize=[15, 7])
data.groupby(by=['Director'])['Revenue (Millions)'].sum().sort_values(ascending=False)[0:10].plot.bar( color="#2ecc71")
plt.xticks(size=12, rotation=45)
plt.yticks(ticks=np.arange(0, 1900, 250), size=12)
plt.xlabel(xlabel='Director', size=14)
plt.ylabel(ylabel='Revenue IN (Millions)', size=14)
plt.title(label='Top 10 Director Made Highest Revenue', size=16)
plt.show()
```

```python
figure = plt.figure(figsize=[15, 7])
data["Genre"].value_counts()[0:10].plot.bar(color='#7d3c98',)
plt.xticks(size=12, rotation=45)
plt.yticks( size=12)
plt.xlabel(xlabel='Genre', size=14)
plt.ylabel(ylabel='Count ', size=14,rotation=90)
plt.title(label='Bar Chart For Top 10 Genre With highest movies ', size=16)
plt.show()
```

# Avarage Revenue Of Movies Per Year

```python
figure = plt.figure(figsize=[15, 7])
data.groupby(by=['Year'])['Revenue (Millions)'].mean().plot.bar( color="#52be80")

plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 140, 20), size=12)
plt.xlabel(xlabel='YEARS', size=14)
plt.ylabel(ylabel='Revenue IN (Millions)', size=14)
plt.title(label='Avarage Revenue Of Movies Per Year', size=16)

plt.show()
```

# Revenue Of Movies Per Year


```python
figure = plt.figure(figsize=[15, 7])
data.groupby(by=['Year'])['Revenue (Millions)'].sum().plot.bar( color="#7d3c98")

plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 140, 20), size=12)
plt.xlabel(xlabel='YEARS', size=14)
plt.ylabel(ylabel='Revenue IN (Millions)', size=14)
plt.title(label='Revenue Of Movies Per Year', size=16)

plt.show()
```

# Revenue Generated By Genre Top 10 

```python
figure = plt.figure(figsize=[15, 7])
data.groupby(by=['Genre'])['Revenue (Millions)'].sum().sort_values(ascending=False)[0:10].plot.bar( color="#196f3d")
plt.xticks(size=12, rotation=45)
plt.yticks(ticks=np.arange(0, 12000, 1000), size=12)
plt.xlabel(xlabel='Genre', size=14)
plt.ylabel(ylabel='Revenue IN (Millions)', size=14)
plt.title(label='Top 10 Profitable Genre', size=16)

plt.show()
```

## Top 10 Actors With highest Revenue

```python
figure = plt.figure(figsize=[15, 7])
data.groupby(by=['Actors'])['Revenue (Millions)'].sum().sort_values(ascending=False)[0:10].plot.bar( color='#a9dfbf')
plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 1200, 200), size=12)
plt.xlabel(xlabel='Actors', size=14)
plt.ylabel(ylabel='Revenue IN (Millions)', size=14)
plt.title(label='Top 10 Actors With highest Revenue', size=16)

plt.show()
```
